```python
import numpy as np
def createMaze1():
    maze = []
    maze.append([*"#####O#"])
    maze.append([*"#   # #"])
    maze.append([*"# # # #"])
    maze.append([*"# #   #"])
    maze.append([*"# ### #"])
    maze.append([*"#   # #"])
    maze.append([*"###X###"])

    start = None
    for i, item in enumerate(maze[0]):
        if item == "O":
            start = (0, i)
    return np.array(maze), start

def createMaze2():
    maze = []
    maze.append([*"#####O###"])
    maze.append([*"#       #"])
    maze.append([*"# ## ## #"])
    maze.append([*"# #   # #"])
    maze.append([*"# # # # #"])
    maze.append([*"# # # # #"])
    maze.append([*"# # # ###"])
    maze.append([*"#       #"])
    maze.append([*"##X######"])
    start = None
    for i, item in enumerate(maze[0]):
        if item == "O":
            start = (0, i)
    return np.array(maze), start

def valid(next_node, maze, visited):
    if next_node[0] < 0 or next_node[1] < 0:
        return False
    if next_node[0] >= maze.shape[0]:
        return False
    if next_node[1] >= maze.shape[1]:
        return False
    
    if next_node in visited:
        return False
    if maze[next_node] == "#":
        return False
    return True

def find_path(maze, start):
    visited = list()

    directions = [(0,1), (1,0), (0,-1), (-1,0)]
    visited.append(start)
    queue = list()
    queue.append((start, []))
    while queue:
        (node, path) = queue.pop(0)
        for dx, dy in directions:
            next_node = (node[0]+dx, node[1]+dy)
            if valid(next_node, maze, visited):
                if maze[next_node] == "X":
                    return path + [next_node]
                else:
                    visited.append(next_node)
                    queue.append((next_node, path + [next_node]))

maze, start = createMaze1()
path = find_path(maze, start)


for p in path:
    maze[p] = "."

for row in maze:
    print("".join(row))

```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from queue import Queue

def create_maze(dim):
    # Create a grid filled with walls
    maze = np.ones((dim*2+1, dim*2+1))

    # Define the starting point
    x, y = (0, 0)
    maze[2*x+1, 2*y+1] = 0

    # Initialize the stack with the starting point
    stack = [(x, y)]
    while len(stack) > 0:
        x, y = stack[-1]

        # Define possible directions
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        random.shuffle(directions)

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if nx >= 0 and ny >= 0 and nx < dim and ny < dim and maze[2*nx+1, 2*ny+1] == 1:
                maze[2*nx+1, 2*ny+1] = 0
                maze[2*x+1+dx, 2*y+1+dy] = 0
                stack.append((nx, ny))
                break
        else:
            stack.pop()
            
    # Create an entrance and an exit
    maze[1, 0] = 2.0
    maze[-2, -1] = 3.0

    return maze

maze = create_maze(100)
def num_to_str(num):
    if num == 1.0:
        return "#"
    if num == 2.0:
        return "O"
    if num == 3.0:
        return "X"
    return " "
    
maze = np.array([[num_to_str(x) for x in row] for row in maze])
start = (1,0)
path = find_path(maze, start)


for p in path:
    maze[p] = "."

for row in maze:
    print("".join(row))
```
